# Stateful LSTM

RNN podem manter seus estados ao longo dos batches no treinamento. O estado oculto calculado para o batch dos dados de treinamento serão utilizados para inicializar os estados ocultos para o seguinte batch. Embora, está carateristica tem que ser configurada, por defeito keras reset os estados ocultos ao final de processar cada batch. Configurar uma RNN para salvar seus estados ocultos atraves do treinamento e manter esta carateristica durante a predição.

Os beneficios de utilizar RNNs são redes neurais de menor dimensão e/ou menor tempo de treinamento. A desvantagem é que o cientista é responsavel por treinar a rede com uma dimensão do batch dependente da periodicidade dos dados, e reiniciar os dados após cada epoca. Também, os dados não podem embaralhados pra a uniformidade, pois o ordem dos dados é relevante para o estado da rede. 

## Exemplo: Previsão do consumo de eletricidade

Neste exemplo, vamos prever o consumo de electricidade de um consumidor e compar o resultado de LSTMs com e sem armazenamento de seu estado oculto. O dataset utilizado é `Electricity Load Diagrams` de [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/ElectricityLoadDiagrams20112014), e contem informação de consumo de 370 clientes, com dados tomados em intervalos de 15 minutos em um periodo de 4 anos desde 2011 até 2014. Foi selecionado o cliente # 250 para nosso exemplo.

Um ponto importante é que a maioria dos problemas podem ser resolvidos utilizando RNN sem manter seu estado ao longo do treinamento dos batchs. Se você vai manter os estados tem que ter certeza de que precisa de ele. Normalmente são utilizados quando os dados tem uma dependencia periodica. Se analisamos um pouco podemos inferir que o consumo de electricidade é periodico. Consumo tem uma tendencia em ser maior durante o dia que na noite. 

# Pre-processamento

Como podemos observar, existe claramente uma tendencia periodica diaria. Asim, o problema é um bom candidato para nosso modelo. Também, baseado em nosso observação, a `BATCH_SIZE` de 96 (# de dados cada 15 minutos para 24 horas) é apropiado. Agora, para apresentar os dois modelos, primeiro vamos importar as bibliotecas e classes necessarias:

```python
from keras.layers.core import Dense
from keras.layers.recurrent import LSTM
from keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
from keras_tqdm import TQDMNotebookCallback
import numpy as np
import math
import os
import matplotlib.pyplot as plt
```

Agora, abrimos os dados do cliente 250 num array de dimensão (140256) do arquivo binario salvado e re-escalamos os valores entre 0-1. Finalmente redimensionamos a entrada para 3 dimensões como precisa nossa rede.

```python
DATA_DIR = "data"

np_file=os.path.join(DATA_DIR,"LD_250.npy")
data = np.load(np_file)

# scale the data to be in the range (0, 1)
data = data.reshape(-1,1)
scaler = MinMaxScaler(feature_range=(0,1), copy=False)
data = scaler.fit_transform(data)
```

```python
NUM_ENTRIES = 1000
plt.plot(range(NUM_ENTRIES), data[0:NUM_ENTRIES])
plt.ylabel("electricity consumption")
plt.xlabel("time (1pt = 15 mins)")
plt.show()
```

Em cada batch, o modelo pegara uma sequencia de 15 minutos de leitura e preveera o seguiente.  A dimensão da sequencia de entrada é dada pela variavel `NUM_TIMESTEPS` no código. Baseados em alguns experimentos realizados, obtimos o valor para `NUM_TIMESTEPS` de 20. Então, cada fila tem uma sequencia de dimensão 20 e a saida de dimensão 1. O seguinte passo reordena as matrizes $x$ e $y$ em tensor de entrara de dimensão `(None,4)` e `(None,1)`. Finalmente, redimensionamos o tensor de entrada $x$ para 3 dimensões para a entrada de nossa rede.

```python
NUM_TIMESTEPS=20

# transform to 4 inputs -> 1 label format
X = np.zeros((data.shape[0], NUM_TIMESTEPS))
Y = np.zeros((data.shape[0],1))
for i in range(len(data) - NUM_TIMESTEPS - 1):
    X[i] = data[i:i + NUM_TIMESTEPS].T
    Y[i] = data[i + NUM_TIMESTEPS + 1]

# reshape X to three dimensions (samples, timesteps, features)
X = np.expand_dims(X, axis=2)
```

Finalmente, dividimos nossos dados de treinamento e teste (70-30). Como trabalhamos com series temporais, escolhems o punto de corte e dividimos a data em 2 partes, e não utilizar a função `train_test_split`, que embaralha os dados.

```python
# split into training and test sets
sp = int(0.7*len(data))
Xtrain, Xtest, Ytrain, Ytest = X[0:sp], X[sp:], Y[0:sp], Y[sp:]
print(Xtrain.shape, Xtest.shape, Ytrain.shape, Ytest.shape)
```

# Criação dos modelos

O primeiro modelo será sem armzenar o estado ao longo do treinamento. Configuramos o `BATCH_SIZE` e `NUM_TIMESTEPS`, como foi detalhado anteriormente. A dimensão de nossa LSTM é dada pela `HIDDEN_SIZE`. Neste caso só configuramos com 10 pois só queremos comparar 2 redes.

```python
HIDDEN_SIZE = 10
BATCH_SIZE = 96 # 24 hours (15 min intervals)
NUM_EPOCHS = 5
STATELESS = True
```

```python
if STATELESS:
    # stateless
    model = Sequential()
    model.add(LSTM(HIDDEN_SIZE, input_shape=(NUM_TIMESTEPS, 1), 
                   return_sequences=False))
    model.add(Dense(1))
else:
    # stateful
    model = Sequential()
    model.add(LSTM(HIDDEN_SIZE, stateful=True,
                   batch_input_shape=(BATCH_SIZE, NUM_TIMESTEPS, 1), 
                   return_sequences=False))
    model.add(Dense(1))

model.compile(loss="mean_squared_error", optimizer="adam",
              metrics=["mean_squared_error"])
```

# Treinamento

```python
if STATELESS:
    # stateless
    model.fit(Xtrain, Ytrain, epochs=NUM_EPOCHS, batch_size=BATCH_SIZE,
              validation_data=(Xtest, Ytest),
              shuffle=False,
              verbose=0, callbacks=[TQDMNotebookCallback()])
else:
    # stateful
    # need to make training and test data to multiple of BATCH_SIZE
    train_size = (Xtrain.shape[0] // BATCH_SIZE) * BATCH_SIZE
    test_size = (Xtest.shape[0] // BATCH_SIZE) * BATCH_SIZE
    Xtrain, Ytrain = Xtrain[0:train_size], Ytrain[0:train_size]
    Xtest, Ytest = Xtest[0:test_size], Ytest[0:test_size]
    print(Xtrain.shape, Xtest.shape, Ytrain.shape, Ytest.shape)
    for i in range(NUM_EPOCHS):
        model.fit(Xtrain, Ytrain, batch_size=BATCH_SIZE, epochs=1,
                  validation_data=(Xtest, Ytest),
                  shuffle=False, verbose=0, callbacks=[TQDMNotebookCallback()])
        model.reset_states()

score, _ = model.evaluate(Xtest, Ytest, batch_size=BATCH_SIZE, verbose=0)
rmse = math.sqrt(score)
print("\nMSE: {:.4f}, RMSE: {:.4f}".format(score, rmse))
```

## Observações (Stateful Model):
**1.-** Seleccionar o valor de `BATCH_SIZE` que represente a periodicidade de seus dados. Isto é porque está RNN relaciona seu estado interno de um batch com o seguinte, pelo qual seleccionar um adequado `BATCH_SIZE` ajudara a RNN aprender de uma forma mais rapida. A dimensão de seus dados de treinamento e teste precisam ser multiplos exatos de `BATCH_SIZE`. <p/>
**2.-** Você precisa treinar o modelo manualmente em um bucle for com o ńumero de epocas que você desejar. cada iteração vai treinar o modelo por 1 epoca, e o estado vai se manter ao longo da epoca. Depois o estado precisa ser resetado manualmente.<p/>
**3.-** A terceira coisa é que nossa data precissa ser fornecida em sequencia (`shuffle=False`).
