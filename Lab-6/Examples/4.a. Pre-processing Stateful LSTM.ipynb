# Pre-processamento: Previsão do consumo de eletricidade
Vamos pegar o consumo do cliente # 250 e mostrar os primeiros 10 dias de dados. Finalmente, armazenaremos num arquivo Numpy binário para os seguintes passos.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import re

DL_DATA="/share/apps/DL_DATA/LSTM"
DATA_DIR = "data"
file = os.path.join(DL_DATA, "LD2011_2014.txt")
np_file=os.path.join(DATA_DIR,"LD_250.npy")

fld = open(file, "rb")
data =[]
cid = 250
for line in fld:
    #line = line.decode("ascii","ignore")
    if line.startswith(b'"\";'):
        continue
    cols = [float(re.sub(b',', b'.',x)) for x in line.strip().split(b';')[1:]]
    data.append(cols[cid])
fld.close()

NUM_ENTRIES = 1000
plt.plot(range(NUM_ENTRIES), data[0:NUM_ENTRIES])
plt.ylabel("electricity consumption")
plt.xlabel("time (1pt = 15 mins)")
plt.show()

np.save(np_file, np.array(data))
```
